## 3. Data Types and Structures

### Explanation
Python has several built-in data types including integers, floats, strings, and booleans. It also has data structures like lists, tuples, sets, and dictionaries.

### Example

```python
# Numeric types
integer_num = 42
float_num = 3.14

# Strings
text = "Python is awesome!"

# Lists
fruits = ["apple", "banana", "cherry"]

# Dictionaries
person = {"name": "Bob", "age": 30, "city": "New York"}

print(f"Integer: {integer_num}")
print(f"Float: {float_num}")
print(f"String: {text}")
print(f"List: {fruits}")
print(f"Dictionary: {person}")
```

## 4. Control Flow

### Explanation
Control flow in Python is managed by conditional statements (if-elif-else) and loops (for and while).

### Example

```python
# If-else statement
age = 18
if age >= 18:
    print("You are an adult")
else:
    print("You are a minor")

# For loop
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(f"I like {fruit}")

# While loop
count = 0
while count < 3:
    print(f"Count is {count}")
    count += 1
```

## 5. Functions

### Explanation
Functions in Python are defined using the `def` keyword. They can take parameters and return values.

### Example

```python
def greet(name):
    return f"Hello, {name}!"

def calculate_area(length, width):
    return length * width

# Using the functions
print(greet("Charlie"))
print(f"The area of a 5x3 rectangle is: {calculate_area(5, 3)}")
```
